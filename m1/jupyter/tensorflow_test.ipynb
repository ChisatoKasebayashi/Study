```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split

sess = tf.Session()

data = np.loadtxt('../data/chisato/goal/log201805261417.csv', delimiter=',', dtype='float')
#print(data.shape)
rx = np.array([data[:,0]])
ry = np.array([data[:,1]])
jx = np.array([data[:,2]])
jy = np.array([data[:,3]])
robo = np.transpose(np.concatenate([rx,ry],axis=0))
joy = np.transpose(np.concatenate([jx,jy],axis=0))
X_train, X_test, Y_train, Y_test = train_test_split(robo,joy,test_size=0.2,random_state=0)

x_data = tf.placeholder(shape=[None, 2],dtype=tf.float32)
y_target = tf.placeholder(shape=[None, 2],dtype=tf.float32)

num_units1 = 2
w2 = tf.Variable(tf.truncated_normal([2, num_units1]))
b2 = tf.Variable(tf.zeros([num_units1]))
hidden1 = tf.nn.relu(tf.matmul(x_data, w2) + b2)
print(hidden1)
num_units2 = 3
w1 = tf.Variable(tf.truncated_normal([num_units1, num_units2]))
b1 = tf.Variable(tf.zeros([num_units2]))
hidden2 = tf.nn.relu(tf.matmul(hidden1,w1) + b1)
print(hidden2)
w0 = tf.Variable(tf.truncated_normal([num_units2, 2]))
b0 = tf.Variable(tf.zeros([2]))
final = tf.nn.relu(tf.matmul(hidden2,w0) + b0)
print(final)
loss = tf.reduce_mean(tf.nn.l2_loss(Y_train-final))
print("Y_train - final:"+ str( tf.reduce_mean(tf.nn.l2_loss(Y_train-final))))
optimizer = tf.train.GradientDescentOptimizer(0.2)
train_step = optimizer.minimize( loss )

init = tf.global_variables_initializer()
sess.run(init)
for i in range(100):
    x_in = np.array([X_train[i]])
    y_in = np.array([Y_train[i]])
    sess.run(train_step, feed_dict = { x_data:x_in, y_target:y_in })
    l = sess.run(loss,feed_dict = {x_data:x_in, y_target:y_in})
    print(l)
```
